# **Executive Summary**

To streamline the home equity line of credit application process, the bank tasked its data science team with creating a predictive model. This model was to serve two main objectives: (1) Predicting loan applicants who were likely to default on their loans. (2) Identifying key features that led to a loan rejections for an applicant. By leveraging this model, the bank aimed to enhance the loan application process while ensuring compliance with the Equal Credit Opportunity Act and fair banking regulations. Customers will receive transparent reasons for loan denials, enabling them to understand the decision-making process. Meanwhile, the bank would benefit from offering loans to creditworthy customers, thereby optimizing its lending practices and mitigating risks associated with defaults.


In credit risk assessment, discerning customers likely to default on their loans from those expected to repay them in full is a classification problem. Various models were available for predicting such outcomes, but in this project, I analyzed five distinct models: decision trees, random forests, logistic regression, AdaBoost, and XGBoost. Decision trees classified customers into defaulters and non-defaulters through a sequence of if-else decisions. Random forests, similar to decision trees, employed multiple decision outcomes to classify customers into defaulters and non-defaulters. Logistic regression, on the other hand, modeled the relationship between probability of defaulting and the features. AdaBoost employed simple models, often referred to as weak learners, to classify customers into defaulters and non-defaulters. Lastly, XGBoost, similar to AdaBoost, incorporated additional techniques to enhance both accuracy and speed in the classification process. Through exploring these diverse models, I compared and selected a model for predicting loan default risk and to facilitate prudent decision-making in the bank's lending practices.

    
Prior to evaluating the aforementioned models, our dataset underwent comprehensive analysis to ensure robustness and accuracy. This encompassed exploratory analysis and correlation analysis to glean insights into the data's structure and relationships. To address the issues of missing values, simple imputation was applied specifically for the REASON and JOB features, while the KNN imputer was utilized to impute the remainder of the features. Moreover, features exhibiting outliers were subjected to capping at 90% of their respective distributions to maintain consistency and minimize skewness. Furthermore, given the inherent class imbalance, where only 20% of the dataset comprised defaulters while 80% constituted non-defaulters, up-sampling was employed to rectify this disproportionality. This maneuver mitigated the potential adverse effects of class imbalance during the modeling process, thereby enhancing the reliability and efficacy of our predictive models.


In our modeling process, I split data into training and testing sets. The training set was used for model training and the testing set was used for evaluating model performance on unseen data. Model performance was evaluated based on two crucial metrics: recall, which indicated the proportion of correctly identified defaulting customers from the total defaulters, and precision, which indicated the proportion of true defaulters among those predicted to default. To tune model parameters, I focused primarily on maximizing recall and precision. I specifically compared these two metrics for the five models (see Table below). From this comparison, random forest model emerged as the best performing model with recall of 74% and precision of 31%. 
A recall of 74% means that out of all the actual defaulting customers in the dataset, the model correctly identified 74% of them as defaulters. A precision of 31% means that out of all the customers predicted by the model to be defaulters, only 31% of them were actually defaulters.

## MOD 5d: Cap outliers

```python
X_train_imp_outliers = X_train_imp.copy()
print( X_train_imp_outliers.shape )

X_test_imp_outliers = X_test_imp.copy()
print( X_test_imp_outliers.shape )
```

```python
def func_outliers (dfin, var, trans):
    
    dsn = dfin.copy()

    if trans== 1:
        # if right skewed then transform variable (make it more normal)
        v1 = np.log(dsn[var]+1)

    if trans==0:
        v1 = dsn[var]

    # set threshold to 2
    thres = 2

    # Computing 10th, 90th percentiles and replacing the outliers 
    ninetieth_per = np.percentile(dsn[var], 90) 

    mean = np.mean(v1)
    std = np.std(v1)
    dsn['z_score'] = (v1-mean)/std
    dsn['outlier_flag'] = 0

    # flag outliers
    dsn['outlier_flag'+'_'+var] = 0
    dsn.loc[ (np.abs(dsn['z_score']>thres)), 'outlier_flag'+'_'+var] = 1 

    # replace outliers with their 90 percentiles
    dsn.loc[ (np.abs(dsn['z_score']>thres)), var] = ninetieth_per 

    print(f"Number of outliers for variable {var} are: {dsn['outlier_flag'+'_'+var].sum() }")

   
    return(dsn)

# training set outliers
df_outliers_train = func_outliers(dfin = X_train_imp_outliers, var = 'LOAN', trans = 1)
df_outliers_train = func_outliers(dfin = df_outliers_train, var = 'MORTDUE', trans = 1)
df_outliers_train = func_outliers(dfin = df_outliers_train, var = 'VALUE', trans = 1)
print("="*100)

# test set outliers
df_outliers_test = func_outliers(dfin = X_test_imp_outliers, var = 'LOAN', trans = 1)
df_outliers_test = func_outliers(dfin = df_outliers_test, var = 'MORTDUE', trans = 1)
df_outliers_test = func_outliers(dfin = df_outliers_test, var = 'VALUE', trans = 1)
 
```
